# Demonstrate gen onboarding name, using onboarding data

```python
from dotenv import load_dotenv
import pandas as pd
from onboarding_llm import llm_onboarding_label
import os

load_dotenv()

LLM_PROVIDER = os.getenv("LLM_PROVIDER", "azure")

INPUT_CSV = "./data/data_raw_input/2025-11-13/CAMPAIGNS.csv"   
OUTPUT_CSV = "./data/output_onboarding/onboarding_names.csv"
MAX_WORDS = 5                        # can override per your requirement

df = pd.read_csv(INPUT_CSV)

def _apply_llm(row):
    label, _raw = llm_onboarding_label(
        row,
        provider=LLM_PROVIDER,
        max_words=MAX_WORDS,
    )
    return label

df["onboarding_name"] = df.apply(_apply_llm, axis=1)

df.to_csv(OUTPUT_CSV, index=False)
print(f"Wrote {len(df)} rows with onboarding_name to {OUTPUT_CSV}")
```
