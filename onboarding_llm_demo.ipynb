```python
import pandas as pd
from onboarding_llm import llm_onboarding_label

INPUT_CSV = "onboarding_data.csv"     # or "onboarding_schema.csv" if you overwrote it
OUTPUT_CSV = "onboarding_names.csv"
LLM_PROVIDER = "openai"              # "openai" or "azure"
MAX_WORDS = 5                        # can override per your requirement

df = pd.read_csv(INPUT_CSV, nrows=5)

def _apply_llm(row):
    label, _raw = llm_onboarding_label(
        row,
        provider=LLM_PROVIDER,
        max_words=MAX_WORDS,
    )
    return label

df["onboarding_name"] = df.apply(_apply_llm, axis=1)

df.to_csv(OUTPUT_CSV, index=False)
print(f"Wrote {len(df)} rows with onboarding_name to {OUTPUT_CSV}")
```
